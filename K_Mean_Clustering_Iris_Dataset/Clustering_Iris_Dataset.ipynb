```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn.metrics as sKm
%matplotlib inline
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
plt.figure(figsize=(13,4))
colors = np.array(['green', 'blue', 'red'])
iris_targets_legend = np.array(iris.target_names)
green = mpatches.Patch(color='green', label='Setosa')
blue = mpatches.Patch(color='blue', label='Versicolor')
red = mpatches.Patch(color='red', label='Virginica')
plt.subplot(1, 2, 1)
plt.scatter(x['Sepal Length'], x['Sepal Width'], c=colors[y['Species']])
plt.title('Sepal Length vs Sepal Width')
plt.legend(handles=[green, blue, red])
plt.subplot(1,2,2)
plt.scatter(x['Petal Length'], x['Petal Width'], c= colors[y['Species']])
plt.title('Petal Length vs Petal Width')
plt.legend(handles=[green, blue, red])
```

```python
dist_from_center = []
K = range(1,10)

for no_of_clusters in range(1,10):
  k_model_ = KMeans(n_clusters=no_of_clusters)
  k_model_.fit(x)
  dist_from_center.append(k_model_.inertia_)
```

```python
dist_from_center
```

```python
plt.plot(K, dist_from_center)
```

```python
k_model = KMeans(n_clusters=3)
k_model.fit(x)
print (k_model.labels_)
```

```python
print (k_model.cluster_centers_)
```

```python
plt.figure(figsize=(13,4))

colors = np.array(['green', 'blue', 'red'])

predict = np.choose(k_model.labels_, [2, 0, 1])

plt.subplot(1, 2, 1)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[y['Species']])
plt.title('Before classification')
plt.legend(handles=[green, blue, red])

plt.subplot(1, 2, 2)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[predict])
plt.title("Model's classification")
plt.legend(handles=[green, blue, red])
```

```python
sKm.accuracy_score(predict, y['Species'])
```

```python
sKm.confusion_matrix(predict, y['Species'])
```
